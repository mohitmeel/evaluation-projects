there are null values present in the dataset so the null values are to be replaced by mean values.

```python
for i in df.columns:
    if df[i].dtypes=='float64':
        np.mean(df[i])
        df[i]=df[i].fillna(np.mean(df[i]))
```

```python
le=LabelEncoder()
for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=le.fit_transform(df[i].values.reshape(-1,1))
```

the important library are called for the modeling and thus predicting the columns. 
the model which we are calling is regression as the target column is the continuous data.

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

splitting the dataset for the machine to learn in the form of train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=90,test_size=.25)
```

```python
x_train.shape
```

finding out the best random state

```python
maxsc=0
maxrs=0
for i in range(1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=i)
    lm=LinearRegression()
    lm.fit(x_train,y_train)
    predlm=lm.predict(x_test)
    score=lm.score(x_train,y_train)
    if score>maxsc:
        maxsc=score
        maxrs=i
print(maxsc)
print(maxrs)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
predlm=lm.predict(x_test)
print('mean absolute error:',mean_absolute_error(y_test,predlm))
print('mean squared error:',mean_squared_error(y_test,predlm))
print(lm.score(x_train,y_train))
print(r2_score(y_test,predlm))
```

```python
ls=Lasso(alpha=.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
print(ls.coef_)
```

```python
rg=Ridge(alpha=.0001)
rg.fit(x_train,y_train)
print(rg.score(x_train,y_train))
print(rg.coef_)
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
predrfr=rfr.predict(x_test)
print('predicted temperature=',predrfr)
print('actual temperature=',y_test)
print('score',rfr.score(x_train,y_train))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=predrfr,color='r')
plt.plot(y_test,y_test,color='b')
plt.xlabel('actual price',fontsize=14)
plt.ylabel('predicted price',fontsize=14)
plt.title('random forest regressor',fontsize=18)
plt.show()
```
