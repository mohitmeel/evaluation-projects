after removing the skewness now the dataset is being divided into target column and feature column as y and x respectively.

```python
x=df.iloc[:,1:]
x
```

```python
y=df.iloc[:,0]
y
```

```python
x.shape
```

```python
y.shape
```

performing the standard scaler method for making the data in the scaler format so the machine can easily learn the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x1=sc.fit_transform(x)
x1
```

```python
x=pd.DataFrame(x1)
x
```

```python
x.shape
```

the important library are called for the modeling and thus predicting the columns. 
the model which we are calling is regression as the target column is the continuous data.

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.svm import SVR
```

splitting the dataset for the machine to learn in the form of train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

the first model which i am using is the linear regression and the score which i got is 97.5

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
predlm=lm.predict(x_test)
print('mean absolute error:',mean_absolute_error(y_test,predlm))
print('mean squared error:',mean_squared_error(y_test,predlm))
print(lm.score(x_train,y_train))
print('r2 score',r2_score(y_test,predlm))
```

```python
df2=pd.DataFrame(print(y_test))
df2
```

now performing the lasso and ridge regression to know whether the dataset is overfit or underfit.
the result is that the dataset has no high bias and no high variance as the score is approximately same as
compared to linear regression

```python
ls=Lasso(alpha=.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
print(ls.coef_)
```

```python
plt.figure(figsize=(7,5))
plt.bar(x.columns,ls.coef_)
```

```python
rg=Ridge(alpha=.0001)
rg.fit(x_train,y_train)
print(rg.score(x_train,y_train))
print(rg.coef_)
```

another model is performed which is support vector regressor to get best score.

```python
kernel_list=['linear','poly','rbf']
for i in kernel_list:
    svr=SVR(kernel=i)
    svr.fit(x_train,y_train)
    print(svr.score(x_train,y_train))
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
predlm=lm.predict(x_test)
print('predicted wins=',predlm)
print('actual wins=',y_test)
print('mean absolute error:',mean_absolute_error(y_test,predlm))
print('mean squared error:',mean_squared_error(y_test,predlm))
print(lm.score(x_train,y_train))
print(lm.coef_)
print('r2 score',r2_score(y_test,predlm))
import pickle
filename='baseballlm.pkl'
pickle.dump(lm,open(filename,'wb'))
```

at last the best score which i get from the model is linear regression (97.5).
