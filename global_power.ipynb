finding the best random state

```python
maxsc=0
maxrs=0
for i in range(1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=i)
    lm=LinearRegression()
    lm.fit(x_train,y_train)
    predlm=lm.predict(x_test)
    score=lm.score(x_train,y_train)
    if score>maxsc:
        maxsc=score
        maxrs=i
print(maxsc)
print(maxrs)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
predlm=lm.predict(x_test)
print('mean absolute error:',mean_absolute_error(y_test,predlm))
print('mean squared error:',mean_squared_error(y_test,predlm))
print(lm.score(x_train,y_train))
print(r2_score(y_test,predlm))
```

```python
ls=Lasso(alpha=.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
print(ls.coef_)
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
predrfr=rfr.predict(x_test)
print('predicted capacity_mw',predrfr)
print('actual capacity_mw',y_test)
print('score',rfr.score(x_train,y_train))

plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=predrfr,color='r')
plt.plot(y_test,y_test,color='b')
plt.xlabel('actual price',fontsize=14)
plt.ylabel('predicted price',fontsize=14)
plt.title('random forest regressor',fontsize=18)
plt.show()
```

importing the important libraries to call the model for making the machine learn 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=.25,random_state=120)
```

```python
x1_train.shape
```

```python
x1_test.shape
```

```python
y1_test.shape
```

```python
y1_train.shape
```

finding out the best random state

```python
maxac=0
maxrs=0
for i in range(1,500):
    x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=.25,random_state=i)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    accuracy=accuracy_score(y_test,predlr)
    if accuracy>maxac:
        maxac=accuracy
        maxrs=i
print(maxac)
print(maxrs)
```

```python
model=[LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier()]
for m in model:
    m.fit(x1_train,y1_train)
    m.score(x1_train,y1_train)
    predm=m.predict(x1_test)
    print('acc_score=',accuracy_score(y1_test,predm))
    print('conf_mat=',confusion_matrix(y1_test,predm))
    print('class_report=',classification_report(y1_test,predm))
```

```python
kernel_list=['linear','poly','rbf']
for i in kernel_list:
    svc=SVC(kernel=i)
    svc.fit(x1_train,y1_train)
    svc.score(x1_train,y1_train)
    predsv=svc.predict(x1_test)
    print('acc=',accuracy_score(y1_test,predsv))
    print('conf=',confusion_matrix(y1_test,predsv))
    print('class=',classification_report(y1_test,predsv))
```

```python
kn=KNeighborsClassifier()
cv=cross_val_score(kn,x1,y1,cv=5)
print(cv)
print('mean=',cv.mean())
print('std=',cv.std())
```

```python
model=[RandomForestClassifier(),AdaBoostClassifier(),GradientBoostingClassifier()]
for m in model:
    m.fit(x1_train,y1_train)
    predm=m.predict(x1_test)
    print('acc_score=',accuracy_score(y1_test,predm))
    print('conf_mat=',confusion_matrix(y1_test,predm))
    print('class_report=',classification_report(y1_test,predm))
```

```python
rf=RandomForestClassifier()
rf.fit(x1_train,y1_train)
predrf=rf.predict(x1_test)
print('predicted primary_fuel=',predrf)
print('actual primary_fuel=',y1_test)
print('accuracy_score=',accuracy_score(y1_test,predrf))
print('confusion_matrix=',confusion_matrix(y1_test,predrf))
print('classification_report=',classification_report(y1_test,predrf))
```

finding the best parameter for random forest classifier

```python
from sklearn.model_selection import GridSearchCV
```

```python
parm={'max_features':['auto','sqrt','log2'],'max_depth':[4,5,6,7,8],'criterion':['gini','entropy']}
gcv=GridSearchCV(RandomForestClassifier(),parm,cv=5,scoring='accuracy')
gcv.fit(x1_train,y1_train)
print(gcv.best_params_)
gcv_pred=gcv.best_estimator_.predict(x1_test)
print(accuracy_score(y1_test,gcv_pred))
```

plotting the roc_curve by using the decision tree classifier

```python
from sklearn.metrics import roc_curve
dtc=DecisionTreeClassifier()
dtc.fit(x1_train,y1_train)
print('predtc',dtc.predict_proba(x1_test)[:,1])
fpr, tpr, thresholds = roc_curve(y1_test,dtc.predict_proba(x1_test)[:,1],pos_label=True)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```
